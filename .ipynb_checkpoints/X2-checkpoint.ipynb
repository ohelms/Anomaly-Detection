### Dependencies

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn import preprocessing
from sklearn.svm import OneClassSVM
import statsmodels.api as sm
plt.style.use('ggplot')
```

### Exploration

```python
df = pd.read_csv('td.csv')
df.head()
```

```python
df.Time = pd.to_datetime(df.Time)
```

```python
df['ticks'] = df.Time.map(lambda x: x.strftime('%Y-%m-%d %H'))
ax = df.plot(x=df['ticks'], y='Bits', color='C1', figsize=(20,10))
ax.set_xticks(list(range(0,1657,50)))
ax.set_xticklabels(df['ticks'][::50])
plt.xticks(rotation=60)
plt.show()
```

### Feature Engineering

```python
#extract hours as a new feature
df['hours'] = df['Time'].dt.hour
#idicate whether the hour is during the daytime
df['day'] = ((df['hours'] >= 6) & (df['hours'] <= 20)).astype(int)
#extract weekday as a new feature
df['weekday'] = df['Time'].dt.dayofweek
#idicate whether the weekday is a weekend
df['weekend'] = (df['weekday'] >= 5).astype(int)
# creation of 4 distinct categories that seem useful (week end/day week & night/day)
df['group'] = df['weekend']*2 + df['day']
#outliers
outliers_fraction = 0.01
```

```python
a = df.loc[df['group'] == 0, 'Bits']
b = df.loc[df['group'] == 1, 'Bits']
c = df.loc[df['group'] == 2, 'Bits']
d = df.loc[df['group'] == 3, 'Bits']

fig, ax = plt.subplots()
a_heights, a_bins = np.histogram(a)
b_heights, b_bins = np.histogram(b, bins=a_bins)
c_heights, c_bins = np.histogram(c, bins=a_bins)
d_heights, d_bins = np.histogram(d, bins=a_bins)
fig.set_figwidth(20)
fig.set_figheight(10)

width = (a_bins[1] - a_bins[0])/6

ax.bar(a_bins[:-1], a_heights*100/a.count(), width=width, label='weeknight')
ax.bar(b_bins[:-1]+width, (b_heights*100/b.count()), width=width, label ='weekday')
ax.bar(c_bins[:-1]+width*2, (c_heights*100/c.count()), width=width, label ='weekendnight')
ax.bar(d_bins[:-1]+width*3, (d_heights*100/d.count()), width=width,  label ='weekendday')

plt.legend()
plt.show()
```

### STL Decomposition

```python
stl = df.copy()
stl.set_index(keys=stl.Time,inplace=True)
stl.head()
```

```python
y = stl['Bits'].resample('h').mean()
```

```python
decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
fig.set_figwidth(20)
fig.set_figheight(20)
plt.show()
```

```python
elements = decomposition.resid.fillna(value=0)
mean = np.mean(elements, axis=0)
sd = np.std(elements, axis=0)
df['a3'] = [ 1 if ((i> (mean + sd*2)) or (i< (mean - sd*2))) else 0 for i in elements]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df['ticks'], df['Bits'], color='C1')
ax.set_xticks(list(range(0,1657,50)))
ax.set_xticklabels(df['ticks'][::50])
plt.xticks(rotation=60)
ax.scatter(np.array(df['ticks'])[np.array(df['a3']==1)], np.array(df['Bits'])[np.array(df['a3']==1)], color='C0')
plt.show()
```

### One Class SVM

```python
# Take useful feature and standardize them 
data = df[['Bits', 'hours', 'day', 'weekday', 'weekend']]
min_max_scaler = preprocessing.StandardScaler()
np_scaled = min_max_scaler.fit_transform(data)
# train one class SVM 
model =  OneClassSVM(nu=0.95 * outliers_fraction) #nu=0.95 * outliers_fraction  + 0.05
data = pd.DataFrame(np_scaled)
model.fit(data)
# add the data to the main  
df['a2'] = pd.Series(model.predict(data))
df['a2'] = df['a2'].map( {1: 0, -1: 1} )
print(df['a2'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df['ticks'], df['Bits'], color='C1')
ax.set_xticks(list(range(0,1657,50)))
ax.set_xticklabels(df['ticks'][::50])
plt.xticks(rotation=60)
ax.scatter(np.array(df['ticks'])[np.array(df['a2']==1)], np.array(df['Bits'])[np.array(df['a2']==1)], color='C0')
plt.show()
```

### Isolation Forest

```python
# Take useful feature and standardize them 
data = df[['Bits', 'hours', 'day', 'weekday', 'weekend']]
min_max_scaler = preprocessing.StandardScaler()
np_scaled = min_max_scaler.fit_transform(data)
data = pd.DataFrame(np_scaled)
# train isolation forest 
model =  IsolationForest(contamination = outliers_fraction)
model.fit(data)
# add the data to the main  
df['a1'] = pd.Series(model.predict(data))
df['a1'] = df['a1'].map( {1: 0, -1: 1} )
print(df['a1'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df['ticks'], df['Bits'], color='C1')
ax.set_xticks(list(range(0,1657,50)))
ax.set_xticklabels(df['ticks'][::50])
plt.xticks(rotation=60)
ax.scatter(np.array(df['ticks'])[np.array(df['a1']==1)], np.array(df['Bits'])[np.array(df['a1']==1)], color='C0')
plt.show()
```
